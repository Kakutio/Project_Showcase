```python
from lyrics_manager import LyricsManager, print_album_info
lyrics_manager = LyricsManager()

artist1_album1 = lyrics_manager.album_paths_dict['rap']['en'][1]
artist1_album2 = lyrics_manager.album_paths_dict['rap']['en'][0]

artist2_album1 = lyrics_manager.album_paths_dict['pop_rock']['en'][0]
artist2_album2 = lyrics_manager.album_paths_dict['pop_rock']['en'][1]


albums = [artist1_album1, artist1_album2, artist2_album1, artist2_album2]

for album in albums:
    print_album_info(album)
    print("\n")

```

```python
from llm_chain import OpenAIConnector

connector = OpenAIConnector()

base_system_prompt = ""
with open("system_prompt.txt", "r") as f:
    base_system_prompt = f.read()

print(base_system_prompt)
```

```python
from lyrics_manager import get_album_tracks, get_clean_lines_from_album, load_album, score_quatrains_random, get_album_representation

artist1_album1_data = load_album(artist1_album1)
artist1_album2_data = load_album(artist1_album2)
artist2_album1_data = load_album(artist2_album1)
artist2_album2_data = load_album(artist2_album2)


# --------------------------------- Artist 1, Album 1
artist1_album1_tracks = get_album_tracks(artist1_album1_data)
artist1_album1_clean_lines = get_clean_lines_from_album(artist1_album1_tracks)
artist1_album1_sample_quatrains = get_album_representation(artist1_album1_clean_lines, 100, 42)

artist1_album1_scored_quatrains_good = score_quatrains_random(artist1_album1_sample_quatrains, 35)
artist1_album1_scored_quatrains_good = "\n".join(artist1_album1_scored_quatrains_good)

artist1_album1_scored_quatrains_bad = score_quatrains_random(artist1_album1_sample_quatrains, 15)
artist1_album1_scored_quatrains_bad = "\n".join(artist1_album1_scored_quatrains_bad)



# ---------------------------- Artist 2, Album 1
artist2_album1_tracks = get_album_tracks(artist2_album1_data)
artist2_album1_clean_lines = get_clean_lines_from_album(artist2_album1_tracks)
artist2_album1_sample_quatrains = get_album_representation(artist2_album1_clean_lines, 100, 42)

artist2_album1_scored_quatrains_good = score_quatrains_random(artist2_album1_sample_quatrains, 35)
artist2_album1_scored_quatrains_good = "\n".join(artist2_album1_scored_quatrains_good)

artist2_album1_scored_quatrains_bad = score_quatrains_random(artist2_album1_sample_quatrains, 15)
artist2_album1_scored_quatrains_bad = "\n".join(artist2_album1_scored_quatrains_bad)

#  ------------------------------ Album 2 of both ---> test set.

artist1_album2_tracks = get_album_tracks(artist1_album2_data)
artist1_album2_clean_lines = get_clean_lines_from_album(artist1_album2_tracks)
artist1_album2_sample_quatrains = get_album_representation(artist1_album2_clean_lines, 100, 42)


artist2_album2_tracks = get_album_tracks(artist2_album2_data)
artist2_album2_clean_lines = get_clean_lines_from_album(artist2_album2_tracks)
artist2_album2_sample_quatrains = get_album_representation(artist2_album2_clean_lines, 100, 42)

test_set = []

samples_artist1 = artist1_album2_sample_quatrains[:5]
for sample in samples_artist1:
    test_set.append(sample)


samples_artist2 = artist2_album2_sample_quatrains[:5]
for sample in samples_artist2:
    test_set.append(sample)


# ------------------------ System Prompts (our two setups)
system_prompt_1 = base_system_prompt.replace("{GROUP_0 EXAMPLES}", artist1_album1_scored_quatrains_good)
system_prompt_1 = system_prompt_1.replace("{GROUP_1 EXAMPLES}", artist2_album1_scored_quatrains_bad)

system_prompt_2 = base_system_prompt.replace("{GROUP_0 EXAMPLES}", artist1_album1_scored_quatrains_bad)
system_prompt_2 = system_prompt_2.replace("{GROUP_1 EXAMPLES}", artist2_album1_scored_quatrains_good)


print("FIRST WE'LL DO THIS: ")
print(system_prompt_1)
print("\n\n\n")
print("THEN WE'LL DO THIS: ")
print(system_prompt_2)
print("\n\n\n")
print("WITH THIS: ")

for quatrain in test_set:
    print("-------")
    print(quatrain)




```

```python
# RUNNING THIS COST ~ 1€/10Min. 

eval_responses = []
for i in range(len(test_set)):
    to_prompt = test_set[i]
    response = connector.send_prompt(system_prompt_1, to_prompt)
    
    separator = f"\n\n\n {i} - {test_set[i]} \n : Response:"
    
    print(separator)
    print(response)
    
    
    
    eval_responses.append(separator)
    eval_responses.append(response)

eval_responses = " ".join(eval_responses)

# EVERY TIME YOU RUN THIS CELL, IT WILL OVERWRITE THE FILE
with open("response_setup_1.txt", "w") as f:
    f.write(eval_responses)
```

```python
# RUNNING THIS COST ~ 1€/10Min. 

eval_responses = []
for i in range(len(test_set)):
    to_prompt = test_set[i]         # system prompt 2 this time
    response = connector.send_prompt(system_prompt_2, to_prompt)

    separator = f"\n\n\n {i} - {test_set[i]} \n : Response:"
    
    print(separator)
    print(response)
    eval_responses.append(separator)
    eval_responses.append(response)

eval_responses = " ".join(eval_responses)

# EVERY TIME YOU RUN THIS CELL, IT WILL OVERWRITE THE FILE
with open("response_setup_2.txt", "w") as f:
    f.write(eval_responses)
```
